Merges the information from Tackles.csv and Games.csv based on the common column gameId.

Computes the correlation matrix for the merged DataFrame.

Visualizes the correlation matrix using Seaborn's heatmap.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Merge relevant information from different DataFrames
merged_df = pd.merge(tackles_df, games_df[['gameId', 'week']], on='gameId', how='left')

# Display the first few rows of the merged DataFrame
print("Merged DataFrame:")
print(merged_df.head())

# Explore the correlation matrix
correlation_matrix = merged_df.corr()

# Use Seaborn to visualize the correlation matrix as a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

```python
import matplotlib.pyplot as plt
from datetime import datetime  # Import the datetime module
import os

def plot_team_movement(tracking_df, plays_df, gameId, playIds, team):
        timestamp = datetime.now().strftime('%Y%m%d%H%M%S')
        folder_name = f'{team}_{gameId}_plots_{timestamp}'
            # Create a folder if it doesn't exist
        if not os.path.exists(folder_name):
                os.makedirs(folder_name)


   
        for playId in playIds:
                # Filter tracking_df based on gameId, playId, and team
                filtered_tracking_df = tracking_df[(tracking_df['gameId'] == gameId) 
                                                        & (tracking_df['playId'] == playId)
                                                        & (tracking_df['club'] == team)]

                # Filter plays_df based on gameId and playId
                filtered_plays_df = plays_df[(plays_df['gameId'] == gameId) & (plays_df['playId'] == playId)]

                # Get the ball carrier's nflId
                ball_carrier_id = filtered_plays_df['ballCarrierId'].iloc[0]

                # Display the result
                # print(filtered_tracking_df)

                # Plot x and y coordinates for each player with 'club'='team'
                for _, player_data in filtered_tracking_df.groupby('nflId'):
                        color = 'g' if player_data['nflId'].iloc[0] == ball_carrier_id else 'b'
                        plt.plot(player_data['x'], player_data['y'], marker='o', linestyle='-', color=color, label=player_data['displayName'].iloc[0])

                # Set labels and title
                plt.xlabel('X Coordinate')
                plt.ylabel('Y Coordinate')
                plt.title(f'{team.upper()} Player Movement During Play {playId} in Game {gameId}')

                # Show legend outside the plot area (adjust the coordinates as needed)
                plt.legend(loc='upper right', bbox_to_anchor=(1.75, 1))

                # Save the plot as a PNG file within the folder
                plt.savefig(os.path.join(folder_name, f'{team}_play_{playId}_game_{gameId}.png'), bbox_inches='tight')

                # Clear the current plot to start fresh for the next iteration
                plt.clf()

# Example usage:

plot_team_movement(tracking_week_1_df, plays_df, 2022091110, distinct_playIds, 'KC')
plot_team_movement(tracking_week_1_df, plays_df, 2022091110, distinct_playIds, 'ARI')

```
